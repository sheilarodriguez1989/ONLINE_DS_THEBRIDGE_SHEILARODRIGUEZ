### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

1. Las ciudades ms pobladas, con los nmeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
# Ordenar por poblacin total y seleccionar las 5 ciudades ms pobladas
df_cities_cleaned = df_cities.dropna(subset=["population_total"])
df_top_cities = df_cities.nlargest(5, "population_total")

# Crear el grfico de barras
plt.figure(figsize=(4, 6))
colores = ["blue", "gray", "red", "green", "brown"]
bars = plt.bar(
    df_top_cities["city"], 
    df_top_cities["population_total"] / 1000,  # Convertir poblacin a miles
    color=colores,
    alpha=0.8
)
#Aadir etiquetas encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, #calcula el centro de la barra en el eje x
        height,                            #vamos a colocar la etiqueta encima de la barra
        f'{height:.3f}',                   #cadena formateada para mostrar el valor de la altura con tres decimales

        ha = "center",                     #texto alineado horizontalmente en el centro de la barra
        va="bottom",                       #texto alineado en la parte inferior del texto, justo encima de la barra
        fontsize = 10,
        color = "black"
    )
# Personalizar el grfico
plt.xlabel("Ciudades", fontsize=12)
plt.ylabel("Poblacin (miles)", fontsize=12)
plt.title("Las 5 ciudades ms pobladas de California", fontsize=14)
plt.xticks(rotation=45, ha="right")  # Girar etiquetas del eje x
plt.ylim(0,5000)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python
# Crear el histograma
plt.figure(figsize=(4, 6))
plt.hist(df_cities["area_total_km2"], bins=50, color="steelblue", alpha=0.7, label="Hist Area")  # Aadir etiqueta para la leyenda

# Personalizar el grfico
plt.xlabel("Superficie " + r"$\mathbf{\mathit{(km^2)}}$", fontsize=14, fontname="Bradley Hand ITC")  # La etiqueta se compone de "Superfice" y el LaTeX: r de rawstring mathbf (negrita), mathit(cursiva) y km^2
plt.ylabel("Nmero de ciudades", fontsize=14, fontname="Bradley Hand ITC")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc="upper right", fontsize=10)  # Leyenda con ttulo y posicin
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
# Crear el scatter plot: rea de agua vs poblacin
plt.figure(figsize=(10, 6))
plt.scatter(
    df_cities["area_water_km2"], 
    df_cities["population_total"]/1000, 
    color="green", 
    alpha=0.7
)

outlier_area= df_cities.nlargest(1, "area_water_km2")  # Seleccionar el valor ms alto en rea de agua
outlier_population =df_cities.nlargest(1, "population_total") # Seleccionar el valor ms alto en population_total

#Coordenadas de los outliers
x_outlier_area = outlier_area["area_water_km2"].values[0]
y_outlier_area = outlier_area["population_total"].values[0] / 1000  # Convertir a miles

x_outlier_population = outlier_population["area_water_km2"].values[0]
y_outlier_population = outlier_population["population_total"].values[0] / 1000  # Convertir a miles

# Aadir flecha desde el outlier de rea de agua hasta "Outliers"
arrow_area = FancyArrowPatch(
    posA=(x_outlier_area, y_outlier_area), 
    posB=(80, 3600),  # Coordenadas del texto "Outliers"
    arrowstyle="<-",
    mutation_scale=20,
    color="black"
)
plt.gca().add_patch(arrow_area)

# Aadir flecha desde el outlier de poblacin hasta "Outliers"
arrow_population = FancyArrowPatch(
    posA=(x_outlier_population, y_outlier_population), 
    posB=(80, 3600),  # Coordenadas del texto "Outliers"
    arrowstyle="<-",  #pongo las flechas del revs para que sealen el outlier
    mutation_scale=20,
    color="black"
)
plt.gca().add_patch(arrow_population)

#Aadir texto
plt.text(80, 3600, "Outliers", fontsize=12, ha="right", va="center", color="black")  # Texto "Outliers"


# Personalizar el grfico
plt.xlabel("rea de agua en la ciudad (km)", fontsize=14)
plt.ylabel("Poblacin (miles de habitantes)", fontsize=14)
plt.ylim(-100, 4100)
plt.title("Poblacin por kilmetro cuadrado de agua", fontsize=16)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
# Eliminar valores NaN de la columna elevation_m
elevation_cleaned = df_cities["elevation_m"].dropna()

# Crear el boxplot de la altura de las ciudades
plt.figure(figsize=(8, 6))
plt.boxplot(elevation_cleaned, vert=False, patch_artist=True, boxprops=dict(facecolor="red"), medianprops=dict(color="black"))
plt.ylabel("Elevacin (metros)", fontsize=12)
plt.title("Diagrama de caja de la altitud de las ciudades de California", fontsize=14)

# Personalizar el diseo del grfico
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

5. Unifico las cuatro grficas en una figura:

```python
fig, axes = plt.subplots(2,2, figsize=(16,12))

#Primer grfico:
df_cities_cleaned = df_cities.dropna(subset=["population_total"])
df_top_cities = df_cities.nlargest(5, "population_total")
colores = ["blue", "gray", "red", "green", "brown"]

bars = axes[0,0].bar(
    df_top_cities["city"], 
    df_top_cities["population_total"] / 1000,  
    color=colores,
    alpha=0.8
)
for bar in bars:
    height = bar.get_height()
    axes[0,0].text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.1f}',

        ha = "center",
        va="bottom",
        fontsize = 10,
        color = "black"
    )

axes[0,0].set_xlabel("Ciudades")
axes[0,0].set_ylabel("Poblacin (miles)")
axes[0,0].set_title("Las 5 ciudades ms pobladas de California")
axes[0,0].tick_params(axis="x", rotation=45)  # Girar etiquetas del eje x
axes[0,0].set_ylim(0,5000)


#Segundo grfico:
axes[0,1].hist(df_cities["area_total_km2"], bins=50, color="steelblue", alpha=0.7, label="Hist Area")  # Aadir etiqueta para la leyenda
axes[0,1].set_xlabel("Superficie " + r"$\mathbf{\mathit{(km^2)}}$", fontsize=14, fontname="Bradley Hand ITC")  
axes[0,1].set_ylabel("Nmero de ciudades", fontsize=14, fontname="Bradley Hand ITC")
axes[0,1].tick_params(axis="x")
axes[0,1].tick_params(axis="y")
axes[0,1].legend(loc="upper right")

#Tercer grfico:


axes[1,0].scatter(
    df_cities["area_water_km2"], 
    df_cities["population_total"]/1000, 
    color="green", 
    alpha=0.7
)

outlier_area= df_cities.nlargest(1, "area_water_km2")  # Seleccionar el valor ms alto en rea de agua
outlier_population =df_cities.nlargest(1, "population_total") # Seleccionar el valor ms alto en population_total

arrow_area = FancyArrowPatch(
    posA=(outlier_area["area_water_km2"].values[0], outlier_area["population_total"].values[0]/1000), 
    posB=(80, 3600),  
    arrowstyle="<-",
    mutation_scale=20,
    color="black"
)

arrow_population = FancyArrowPatch(
    posA=(outlier_population["area_water_km2"].values[0], outlier_population["population_total"].values[0]/1000), 
    posB=(80, 3600),  # Coordenadas del texto "Outliers"
    arrowstyle="<-",
    mutation_scale=20,
    color="black"
)

# Personalizar el grfico
axes[1,0].add_patch(arrow_area)
axes[1,0].add_patch(arrow_population)
axes[1,0].text(80,3600, "Outliers", ha="right", va="center", color="black")
axes[1,0].set_xlabel("rea de agua en la ciudad (km)")
axes[1,0].set_ylabel("Poblacin (miles de habitantes)")
axes[1,0].set_title("Poblacin por kilmetro cuadrado de agua")

#Cuarto:

elevation_cleaned = df_cities["elevation_m"].dropna()
axes[1,1].boxplot(
    elevation_cleaned, 
    vert=False, 
    patch_artist=True, 
    boxprops=dict(facecolor="red"), 
    medianprops=dict(color="black")
)
axes[1,1].set_xlabel("Elevacin (metros)")
axes[1,1].set_title("Diagrama de caja de la altitud de las ciudades de California")

#Ajustes de diseo
plt.tight_layout()
plt.savefig("figura_unificada.png", dpi=300)

plt.show()
```
