Vamos a analizar las medidas de tendencia central (media, mediana), las de posicin (los cuartiles) y los rangos de las variables numricas del dataset. Muestra esos valores para todas las variables numricas del dataset.

```python
# Seleccionar las columnas numricas del dataframe
variables_numericas = df_tips.select_dtypes(include=['float64', 'int64'])

# Calcular las medidas de tendencia central y posicin
resultados = variables_numericas.describe()

# Aadir mediana a los resultados (no est incluida por defecto en describe())
resultados.loc["Mediana"] = variables_numericas.median()

# Mostrar resultados
print("Medidas de tendencia central, posicin y rangos:")
print(resultados)
```
