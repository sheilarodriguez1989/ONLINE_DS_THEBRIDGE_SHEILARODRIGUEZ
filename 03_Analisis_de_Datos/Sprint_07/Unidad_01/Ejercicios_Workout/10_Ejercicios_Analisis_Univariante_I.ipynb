### Ejercicio 10

Vamos a analizar las medidas de tendencia central (media, mediana), las de posición (los cuartiles) y los rangos de las variables numéricas del dataset. Muestra esos valores para todas las variables numéricas del dataset.

```python
# Seleccionar las columnas numéricas del dataframe
variables_numericas = df_tips.select_dtypes(include=['float64', 'int64'])

# Calcular las medidas de tendencia central y posición
resultados = variables_numericas.describe()

# Añadir mediana a los resultados (no está incluida por defecto en describe())
resultados.loc["Mediana"] = variables_numericas.median()

# Mostrar resultados
print("Medidas de tendencia central, posición y rangos:")
print(resultados)
```
